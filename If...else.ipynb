## If Statement

if statement is written using the " if " keyword

```python
a = 10
if a == 10:
    print("Hello")
```

```python
a = 5
if a == 6:
    print("Hello")
   

# skip the conditional code if condition is false
# above code doen't print anything
```

if you have only one statement to execute , you can put it on same line as the if statement.

```python
a = 10
if a == 10: print("hello")
```

## Indentation

Indentation is whitespaces at the beginning of the line.

Indentation is use to define the scope of code in python.

if statement without indentation will raise the error.

```python
a = 10
if a == 10:
    print("hello")
    print("Tech Blooded")
```

```python
if a != 10:
    print("tech")
    print("scope of code")  # scope of if is till here

print("Print anyway")
    
```

## else

else keyword catches anything which isn't caught by preceding conditions.

```python
a == 10
if a == 5:
    print("True")
else:
    print("False")
```

```python
if a == 10:
    print("True")
else:
    print("False")
```

### Short hand if...else

if you only have one statement to execute for if and only one for else then you can put it all on same line

```python
print("True") if a == 10 else print("False")
```

## elif 

meaning of elif is "if previous conditions were not true , then try this condition."

```python
a = 10
if a == 5:
    print("1st")
elif a == 6:
    print("2nd")
elif a == 10:
    print("3rd")
```

```python
if a == 2:
    print("a")
elif a == 3:
    print("b")
else:
    print("c")
```

## Nested if

if statement inside if statement is called nested if

```python
a = 10
if a > 5:
    if a == 10:
        print("True")
    else:
        print("False")
```

## Pass statement

if statement cannot be empty , but if your if statement is with no content because of some reasons then you can use "pass"statement to avoid getting error.

```python
a = 10
if a > 5:
    pass
```

```python
if a > 5:
    pass
else:
    print("true")
    
# pass only pass the conditional code.

# above code not giving any output because if condition is true and 
# it passes the conditional code
```
