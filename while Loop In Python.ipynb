# While Loop

By using **while** loop we can execute a set of statements as long as **condition is true**.

### Syntax of while loop

**while** condition :

    Statement or group of statements to execute

```python
i = 1
while i < 11:
    print(i)
    
    i += 1            
    # i = i + 1
```

we need to define the *indexing variable* i

### Break Statement

We can stop the execution of loop whenever we want if the condition is true

```python
i = 1
while i < 11:
    if i == 6:
        break
    print(i)
    
    i += 1
        
```

### Continue Statement

By using continue statement we can **stop the current iteration** and **continue** with the next iteration

```python
i = 0
while i < 10:
    
    i += 1
    
    if i == 5:
        continue
        
    print(i)
    
   
```

### Else Statement

we can run a block of code once when the condition is no longer True

```python
i = 1
while i < 11:
    print(i)
    i+=1
else:
    print("condition no longer true")
```

```python
#Infinity
i = 1
while i < 11:
    print(i)
```
